Question: Why is there a difference between population statistics and sample statistics
    Example: standard deviation formula : population(n) and sample (n-1)
            we divide by a smaller number in the case of a sample 
            because we probably don't capture the outliers in the sample

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
iris = load_iris()
```

```python
iris
```

```python
data = iris.data
labels =iris.target
columns = iris.feature_names

df =pd.DataFrame(data, columns = columns)
df['label'] = labels
df.head()
```

```python
type(df.columns),df.columns
```

```python
#let's rename columns. 
#remove the cm
# add underscores
df.columns = df.columns.str.replace('\(cm\)', '').str.strip().str.replace(" ",'_')
```

```python
df.head()
```

```python
# create a new column with a ratio length /width
# round
df['sepl_sepw_ratio'] = np.round(df['sepal_length']/df['sepal_width'],2)
```

```python
df.head()
```

```python
# new column = distance width to width max
df['petl_from_max'] = np.abs(df['petal_width'].max() - df['petal_width'])
```

```python
df.head()
```

```python
# let's look at our data. histograms
def plot_hist(df,column,figsize=(8,5),bins=20):
    plt.figure(figsize=figsize)
    plt.grid(linestyle = 'dashed', linewidth = 1, zorder=0)
    fontsize=14
    plt.hist(df[column],color ='purple',bins=bins,alpha=0.8,zorder=2)
    plt.title(column.replace("_"," ").title(),fontsize=fontsize)
    plt.xlabel("Values", fontsize = fontsize)
    plt.xlabel("Counts",fontsize = fontsize)
    plt.show()
```

```python
plot_hist(df,'petal_width')
```

```python
for column in df.columns:
    plot_hist(df,column)
```
